```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError

def check_website(url):
    try:
        response = urlopen(url)
        print(f"Страница '{url}' доступна. Код ответа: {response.status}")
    except HTTPError as e:
        print(f"HTTP ошибка: {e.code} для URL: {url}")
    except URLError as e:
        print(f"Ошибка URL: {e.reason} для URL: {url}")
    except Exception as e:
        print(f"Неизвестная ошибка: {e} для URL: {url}")

# Укажите URL, который нужно проверить
url_to_check = "https://www.example.com"
check_website(url_to_check)

# или

from urllib.request import urlopen
from urllib.error import HTTPError, URLError

url = "https://www.python.org"

try:
    response = urlopen(url)
    print(f"Страница найдена: {url}")
except HTTPError as e:
    print(f"HTTP Error: {e.code} - {e.reason}")
except URLError as e:
    print(f"URL Error: {e.reason}")
```

```python
import requests

url = "https://www.python.org"
response = requests.get(url)

# Проверка SSL
if response.url.startswith("https://"):
    print(f"Сайт {url} имеет SSL-сертификат.")
else:
    print(f"Сайт {url} не имеет SSL-сертификата.")

# Вывод параметров страницы
print("Status Code:", response.status_code)
print("Headers:", response.headers)
print("URL:", response.url)
print("History:", response.history)
print("Encoding:", response.encoding)
print("Reason:", response.reason)
print("Cookies:", response.cookies)
print("Elapsed:", response.elapsed)
print("Request:", response.request)
print("Content:", response.text[:500])  # Первые 500 символов контента
```

```python
import random
import requests
from bs4 import BeautifulSoup

# Ссылка на IMDb
imdb_url = "https://www.imdb.com/chart/top"
response = requests.get(imdb_url)

# Проверяем успешность запроса
if response.status_code != 200:
    print("Ошибка: не удалось загрузить страницу IMDb.")
else:
    # Парсинг HTML
    soup = BeautifulSoup(response.text, "html.parser")
    movies = soup.select("td.titleColumn")

    # Проверка количества фильмов
    if len(movies) < 10:
        print(f"На странице доступно только {len(movies)} фильмов.")
        random_movies = movies  # Выбираем все фильмы, если их меньше 10
    else:
        random_movies = random.sample(movies, 10)

    # Вывод случайных фильмов
    print("Список случайных фильмов:")
    for movie in random_movies:
        title = movie.a.text
        year = movie.span.text.strip("()")
        print(f"{title} ({year})")
```

```python
Задача 10 - SOAP
Общее описание задачи - Продвинутый уровень
Требуется реализовать программу, которая будет получать данные при помощи методов любого SOAP сервиса в данном списке. Обратите внимание, что для каждого сервиса есть несколько POST методов. То есть ваша программа должна полностью утилизировать весь функционал выбранного вами SOAP сервиса. Например, в блоке Continents есть всего один метод - CountryInfoService.wso, но во многих других блоках методов больше. 

Язык программирования, библиотеки и паттерны решения на ваш выбор.

При полном выполнении общего описания задачи делать упражнения из Частного описания задачи не требуется. Если вы не уверены, что можете выполнить продвинутый уровень, то прошу перейти к Частному описанию задачи.
Частное описание задачи - Начальный уровень
Здесь мы будем рассматривать ту же базовую постановку, что и в базовом уровне, но посмотрим на перечисленные в ней SOAP сервисы с точки зрения готовых решений и инструментов.
1. Сделать POST запрос при помощи Python с модулем requests
Здесь вам нужно выбрать один из понравившихся методов из списка и попробовать реализовать его.

Например (в примере используются другие сервисы, их использовать нельзя, только те, что в списке выше или другие подобные, которые вы можете по желанию найти в интернете):
import requests
# SOAP URL
url = "http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
 
# structured XML
payload = """<?xml version=\"1.0\" encoding=\"utf-8\"?>
            <soap:Envelope xmlns:soap=\"http://schemas.xmlsoap.org/soap/envelope/\">
                <soap:Body>
                    <CountryIntPhoneCode xmlns=\"http://www.oorsprong.org/websamples.countryinfo\">
                        <sCountryISOCode>RU</sCountryISOCode>
                    </CountryIntPhoneCode>
                </soap:Body>
            </soap:Envelope>"""
# headers
headers = {
    'Content-Type': 'text/xml; charset=utf-8'
}
# POST запрос
response = requests.request("POST", url, headers=headers, data=payload)
 
# ответ
print(response.text)
print(response)

2. Сделать запрос при помощи zeep
Установите zeep при помощи pip install zeep.

Далее попробуйте изучить документацию.
Задайте WSDL URL. В примере я буду использовать
http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL 
http://webservices.oorsprong.org/websamples.countryinfo/CountryIntPhoneCode
http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso
Если вы попробуйте URL выше, то вы можете столкнуться с ошибкой. Для получения правильного WSDL URL можно, просто посетив базовый URL и взять там нужный Service Description. Ошибка сделана специально, чтобы вы попробовали посмотреть документацию сервиса из примера, если вы будете пользоваться приведённым ниже кодом. Выясните, какая из ссылок не работает и почему. Разберитесь в документации и найдите там подходящий метод. Вероятно это что-то с названием …PhoneCode…. Найдите нужные WSDL URL. В качестве базового URL будет использоваться service_url, а после базового URL добавьте имя сервиса.
Далее необходимо создать элемент заголовка и установить в элемент заголовка method_url и service_url.
Инициализируйте zeep с WSDL URL.
Вызовите сервис zeep с именем сервиса. В примере ниже имя сервиса - CountryIntPhoneCode. Необходимо передать параметры с country_code, а также передать заголовок в _soapheaders в виде списка.
Получите телефонный код страны


Пример:


import zeep
 
# WSDL URL
wsdl_url = "http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL"
 
# URL метода
method_url = "http://webservices.oorsprong.org/websamples.countryinfo/CountryIntPhoneCode"
 
# URL сервиса
service_url = "http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso"
 
# создание header element
header = zeep.xsd.Element(
    "Header",
    zeep.xsd.ComplexType(
        [
            zeep.xsd.Element(
                "{http://www.w3.org/2005/08/addressing}Action", zeep.xsd.String()
            ),
            zeep.xsd.Element(
                "{http://www.w3.org/2005/08/addressing}To", zeep.xsd.String()
            ),
        ]
    ),
)
# установка header value из header element
header_value = header(Action=method_url, To=service_url)
 
# инициализация zeep клиента
client = zeep.Client(wsdl=wsdl_url)
 
# ставим код страны Russia
country_code = "RU"
 
# делаем запрос в сервис
result = client.service.CountryIntPhoneCode(
    sCountryISOCode=country_code,
    _soapheaders=[header_value]
)
# выводим результат
print(f"Телефонный код для страны {country_code} - {result}")
 
# вводим код страны США
country_code = "US"
 
# делаем запрос к сервису
result = client.service.CountryIntPhoneCode(
    sCountryISOCode=country_code,
    _soapheaders=[header_value]
)
 
# POST
response = client.service.CountryIntPhoneCode(
    sCountryISOCode=country_code,
    _soapheaders=[header_value]
)
 
# выводим результат
print(f"Телефонный код для страны {country_code} - {result}")
print(response)

```
